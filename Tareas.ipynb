#Tarea 4

Este programa es una calculadora la cual te da los valores de una suma, una resta, una multiplicacion y una divicion se repetira el ciclo hasta que preciones 5 en operaciones

```
menu ="""
Menu
1-SUMA
2-RESTA
3-MULTIPLICACION
4-DIVICION
5-M_salir del programa
"""
pop=True
while pop == True:
    print(menu)
    op = int(input("Escoje una operacion:  "))
    x = float(input(" indica el primer valor: "))
    y = float(input(" indica el segundo valor: "))
    if op is 1:
       print("Has seleccionado suma")
       resultado= x+y
       print("El resultado es")
       print(resultado)  
    elif op is 2:
       print("Has seleccionado resta")  
       resultado= x-y
       print("El resultado es")
       print(resultado)  
    elif op is 3:
       print("Has seleccionado producto") 
       resultado= x*y
       print("El resultado es")
       print(resultado)
    elif op is 4 and y!=0:
       print("Has seleccionado divicion")
       print("Si introduces 0 me regreso al principo")
       resultado= x/y
       print("El resultado es")
       print(resultado)
    elif op is 5:
       pop = False
       print("se acabo se acabo el programa")       
    else:
       print("tecla erronea") 
```

#Tarea 8 
este programa te da un numero de caracteres dependiendo la cantidad que ingreses

```
def triangulo(n):
    return "\n".join(["*"*i for i in range(1,n+1)])
numero=int(input("indica un numero: "))
print(triangulo(numero))
```

#Tarea 10
este prgrama calcula el area de un circulo y el volumen de un cilindro

```
radio= float(input("Cual es el radio del circulo: "))
altura= float(input("Cual es la altura del cilindro: "))
pi=3.1416
area=pi*radio**2
volumen=area*altura
print("El area del circulo es: "+str(area),"cm cuadrados")
print("El volumen del circulo es: "+str(volumen),"cm cubicos")
```
